```python
# 2)

numero = input("Escreva um número: ")
print(f"O número escolhido foi {numero}")
```

```python
# 3)

num1 = float(input("Escreva o primeiro número: "))
num2 = float(input("Escreva o segundo número: "))
resultado = num1 + num2
print("A soma dos dois números é igual a:",resultado)
```

```python
# 4)

nota1 = float(input("Primeira nota bimestral: "))
nota2 = float(input("Segunda nota bimestral: "))
nota3 = float(input("terceira nota bimestral: "))
nota4 = float(input("Quarta nota bimestral: "))

mediaBimestre = (nota1 + nota2 + nota3 + nota4) / 4
print("Sua nota foi igual a", mediaBimestre)
if mediaBimestre != 5.9:
  print("Você passou de Bimestre!!")
else:
  print("Você não passou de Bimestre.")
```

```python
# 5)

medida = int(input("Escreva o tamanho da medida:"))
print("1-Converter de metros para cm")
print("2-Converter de metros para Km")
conversao = input("Digite qual conversão deseja:")

if conversao == "1":
  metro_cm = medida * 100
  print(f"{medida} metros para centimetros é igual {metro_cm} ")

elif conversao == "2":
  metro_km = medida / 1000
  print(f"{medida} metros para quilometros é igual {metro_km}")
```

```python
# 6)

raio = float(input("Qual o raio do círculo? "))
areaCirculo = 3.14 * (raio**2)

print(f"A área da círculo é igual a {areaCirculo}")
```

```python
# 7)

ladoquadrado = float(input("Qual a medida do lado? "))
areaQuadrado = ladoquadrado**2
dobrodaarea = (ladoquadrado**2) * 2
print(f"A área do quadrado é igual a {areaQuadrado}, o dobro da área do quadrado é igual a {dobrodaarea}")
```

```python
# 8)

salario_hora = float(input("O quanto você recebe por hora? "))
horas_trabalho = float(input("Quantas horas você trabalha por dia? "))
dias_mes = float(input("Quantos dias no mês você trabalha? "))
salario_final = salario_hora * horas_trabalho * dias_mes
print(f"Parabéns colega, o seu salário mensal é igual a {salario_final}")
```

```python
# 9)

Fahrenheit = int(input("Qual a temperatura em Fahrenheit? "))
Celsius = int(5 * ((Fahrenheit-32) / 9))

print(f"Essa temperatuda de Fahrenheit em Celsius é aproximadamente {Celsius}°C")
```

```python
# 10)

Celsius = int(input("Qual a temperatura em Celsius? "))
Fahrenheit = int((Celsius * 9 / 5) + 32)

print(f"Essa temperatuda de Celsius em Fahrenheit é aproximadamente {Fahrenheit}°F")
```

```python
# 11)

valor1 = int(input("Escrava o primeiro número inteiro: "))
valor2 = int(input("Escrava o segundo número inteiro: "))
valor3 = float(input("Escrava um número real: "))

resposta1 = (2 * valor1) * (valor2 / 2)
resposta2 = (3 * valor1) + valor3
resposta3 = valor3 ** 3

print(f"O produto do dobro do primeiro com metade do segundo é igual a {resposta1}")
print(f"A soma do triplo do primeiro com o terceiro é igual a {resposta2}")
print(f"O terceiro elevado ao cubo é igual a {resposta3}")
```

```python
# 12)
#Elabore um código que receba os dados de entrada a altura de uma
#pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58
altura = float(input("Qual a sua altura? "))
peso_ideal = int((72.7 * altura) - 58)
print(f"O peso ideal para a sua altura é igual a {peso_ideal}")
```
