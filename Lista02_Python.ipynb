```python
# 19. Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10. A função deve retornar "Aprovado" se a nota for maior ou igual a 7,
#"Recuperação" se estiver entre 5 e 6.9, e "Reprovado" se for menor que 5. Chame a função passando uma nota e imprima o retorno.

def classifica_nota(nota):
  if not (0 <= nota <= 10):
    print("Nota inválida. Por favor, insira um valor entre 0 e 10.")
  elif (nota >= 7):
    print("Aprovado")
  elif (nota >= 5):
    print("Recuperação")
  else:
    print("Reprovado")
```

```python
classifica_nota(10)
```

```python
# 20. Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros). Utilize um laço for com estruturas condicionais
# if/elif/else para contabilizar quantos números são positivos, quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.


lista_de_numeros = [1, 2, 0, -4, 0, -54, -9, -0, 3498]

positivos = 0
negativos = 0
zero = 0

for lista_de_numeros in lista_de_numeros:
  if lista_de_numeros > 0:
    positivos += 1
  elif lista_de_numeros < 0:
    negativos += 1
  else:
    zero += 1

print("Números")
print(f"-Postivos: {positivos}")
print(f"-Negativos: {negativos}")
print(f"-Zeros: {zero}")
```
